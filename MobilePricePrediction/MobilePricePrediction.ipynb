```python
## i decided to predict price range by using the features i assume that they are the most 
## important criteria in the price range.
## in order to that i will follow some steps
## firslt my problem is supervised learning becuse i have labeled data and i will predicted
## my problem is classification beacuse my output is not continous 
## i will predict the range of prices 0,1,2,3
dfVisual=df[['battery_power','int_memory','ram','touch_screen','three_g','price_range']]
## i decided to use these columns 
## battery memory ram screen and 3g as features and price range as a target
```

```python
Featurers = ['battery_power','int_memory','ram','touch_screen','three_g']
target = ['prince_range']
X = dfVisual[Featurers].values
X
## i indicate my featues and target and make them matrix and vector by using values
```

```python
y = dfVisual['price_range'].values
y
```

```python
X2=X ## i create new x2 and y2 in case of something would go wrong
y2=y
```

```python
minX = X2.min(axis=0)   ## applying normalization in order to get more reliable values
maxX = X2.max(axis=0)    ## finding max and min number in a attributes

X2 = (X2-minX)/ (maxX-minX) ## adjusting new values
```

```python
X2
```

```python
from sklearn.model_selection import train_test_split
## %80 train  %20 test set splitting
X_train, X_test, y_train, y_test = train_test_split(X2,y2,test_size=0.2,random_state=42)
## splitting test test two halves %50 test %50 validation 
## in total %80 train %10 test %10 validation
X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.5,random_state=42)
```

```python
from sklearn import svm ## this is my first model which is support vector machine
```

```python
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
clf.score(X_valid, y_valid) ## the score of svm is .805 in my validation set
## and i will try to use decision tree model
## it would give a lower accuracy but it is more intuative 
## when you tell the story about the data to people (which component has a what kind of effect )
```

# Decision Tree Model w/ Default Parameter

```python
from sklearn.tree import DecisionTreeClassifier
## i decided to use Decision Tree because i have a supervised problem that means 
## i will use labeled data and predict some thing
## and my outpul will be discrete in more specific class type
## my target would be 0 1 2 
dt = DecisionTreeClassifier() ## creating model

dt = dt.fit(X_train,y_train) ## fitting it with train sets
dt
```

```python
print(f'test score of trained data of model :{dt.score(X_valid,y_valid)}')
## i test my model on validation set
```

```python
from sklearn import tree ## in order to visualize what the tree is look like

plt.figure(figsize=(30,24))
tree.plot_tree(dt,
              rounded=True, filled=True)
plt.show()
```

# HYPER PARAMETER TUNING

## Grid Search Method

```python
from sklearn.model_selection import GridSearchCV
## i imported gridsearch becuase of the hyperparameter tuning 
## it will provide me to find optimal hyper parameters
## and ranges will be depth 1 to 30 samples split 2 to 10 and sample leaf 2 to 10
parameters = {'max_depth' : range(1,30),
             'min_samples_split' : range(2,10),
             'min_samples_leaf' : range(2,10)}
gcv = GridSearchCV(DecisionTreeClassifier(),parameters).fit(X_train,y_train)
```

```python
pd.DataFrame(gcv.cv_results_) ## results
```

```python
gcv.best_estimator_ ## best model
```

```python
gcv.best_params_ ## best parameters
```

```python
gcv.best_estimator_.score(X_valid,y_valid)

## the first validation set score was .715 and i increased it to .77
##test score of trained data of iris model :0.715
```

```python
print(f'initial model succes {dt.score(X_test,y_test)}')
print(f'optimal model succes {gcv.best_estimator_.score(X_test,y_test)}')

## there is a significant difference between initial and optimal model
## in order to increase model performance i think other classification algorithm would be applied
## or different feature selection scenerios would work better
## or train set may become bigger
```
